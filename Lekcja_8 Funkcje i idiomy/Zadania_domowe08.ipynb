Zadanie 9 – Kategorie BMI dla listy pacjentów

Masz dane pacjentów: wagi = [70, 85, 62, 95, 78] (kg) i wzrosty = [1.75, 1.80,
1.65, 1.78, 1.72] (m). Oblicz BMI dla każdego i sklasyfikuj (niedowaga < 18.5, norma
18.5-24.9, nadwaga >= 25).
Dataset:
Wymagania:
Oczekiwany rezultat:
(średnie)
Listy równoległe: wagi i wzrosty
Użyj pętli for i in range(len(wagi)):
Oblicz bmi = wagi[i] / (wzrosty[i] ** 2)
Sklasyfikuj używając if/elif/else
Wyświetl wyniki dla każdego pacjenta

```python
wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72]

for i in range(len(wagi)):
    BMI = wagi[i]/wzrosty[i]**2
    if BMI < 18.5:
        kat = 'niedowaga'
    elif BMI < 25:
        kat = 'norma'
    else:
        kat = 'nadwaga'
    print(f"waga: {wagi[i]}, wzrost: {wzrosty[i]:.2f}, BMI  {BMI:.2f}, {kat:<9}")
```

Zadanie 11 – Analiza trendów wzrostowych/spadkowych

Masz dane sprzedaży miesięcznej: [1200, 1350, 1290, 1480, 1520, 1510, 1680,
1720, 1700, 1890, 1950, 2100] . Zidentyfikuj miesiące, w których sprzedaż wzrosła w
porównaniu do poprzedniego miesiąca.
Dataset:
Wymagania:
Oczekiwany rezultat:
(średnie)
Lista 12 wartości (sprzedaż miesięczna w PLN)
Pętla for i in range(1, len(sprzedaz)):
Porównaj sprzedaz[i] z sprzedaz[i-1]
Oblicz zmianę procentową
Wyświetl miesiące ze wzrostem

```python
sprzedaz = [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100]

for i in range(1, len(sprzedaz)):
    if sprzedaz[i] > sprzedaz[i-1]:
        wzrost_procentowy = ((sprzedaz[i] - sprzedaz[i-1]) / sprzedaz[i-1]) * 100
        print(f"{i+1:>2}:  wzrost procentowy w porownaniu do poprzedniego miesiaca:{wzrost_procentowy:>7.2f}%")
```

Zadanie 10

Znajdź wszystkie liczby pierwsze do 100
Napisz program, który znajdzie wszystkie liczby pierwsze od 2 do 100 używając
zagnieżdżonych pętli.
Wymagania:
Oczekiwany rezultat:
(średnie)
Pętla zewnętrzna: for n in range(2, 101):
Pętla wewnętrzna: sprawdź dzielniki od 2 do √n
Zmienna czy_pierwsza = True
Jeśli znajdziesz dzielnik, ustaw czy_pierwsza = False i użyj break
Jeśli czy_pierwsza == True , dodaj do listy
Lista liczb pierwszych: [2, 3, 5, 7, 11, 13, ..., 97]
Liczba znalezionych: 25

```python
liczby_pierwsze = []

for i in range(2,101):
    czy_pierwsza = True
    for dz in range(2,i):
        if i % dz == 0:
            czy_pierwsza = False
            break
    if czy_pierwsza == True:
        liczby_pierwsze.append(i)

print(f"Liczby pierwsze z przedzialu 1-100 to {liczby_pierwsze}")
```
